# EDEF Linear Regression Quickstart

EDEF decomposes realized fit, not predictions.

For regression, EDEF explains reductions in realized squared-error loss relative to an intercept-only baseline.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import edef
```

```python
rng = np.random.default_rng(123)

X = rng.normal(size=(300, 3))

beta = np.array([1.0, 0.5, 0.0])

y = X @ beta + rng.normal(scale=0.75, size=300)

feature_names = ["x1_signal", "x2_signal", "x3_noise"]
```

```python
model = LinearRegression()
model.fit(X, y)

explainer = edef.LinearExplainer(model, feature_names=feature_names)

result = explainer(X, y)
```

```python
print(result)
```

```python
result.to_frame()
```

```python
ax = result.plot()
ax.set_title("EDEF contributions to realized fit")
plt.show()
```

Positive EDEF values indicate reductions in realized squared-error loss.

Negative values indicate features that reduced realized fit in this evaluation sample.


```python
grouped = result.group(["signal", "signal", "noise"])

print(grouped)
```

```python
ax = grouped.plot()
ax.set_title("Grouped EDEF contributions")
plt.show()
```

Unlike prediction attribution methods such as SHAP, EDEF depends on realized outcomes `y`.

EDEF explains predictive performance rather than predictions themselves.
